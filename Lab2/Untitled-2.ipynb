```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import os
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/DAVE3625/DAVE3625-24H/refs/heads/main/Lab2/data/Titanic.csv", sep=',')
df.head()
```

```python
df.isna().sum()
```

```python
meanAge = df['Age'][df['Age'].notna()].mean()
meanAge
```

```python
df['Age'].fillna(meanAge, inplace=True)
```

```python
meanFare = df['Fare'][df['Fare'].notna()].mean()
df['Fare'].fillna(meanFare, inplace=True)
```

```python
df['Embarked'].fillna("S", inplace=True)
df.isna().sum()
```

```python
# mean age
print('Mean of "Age" is %.2f' %(df["Age"].mean(skipna=True)))
# median age
print('Median of "Age" is %.2f' %(df["Age"].median(skipna=True)))
```

```python
df['HasCabin'] = ~df.Cabin.isnull()
df.head()
```

```python
df['Title'] = df.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
df
```

```python
df['Title'].value_counts()
```

```python
df["Title"] = df["Title"].replace({'Mlle':'Miss', 'Ms':'Miss', 'Mme':'Mr'})
df['Title'].value_counts()
```

```python
df["Title"] = df["Title"].replace(["Dona","Don","Jonkheer","Rev","Dr","Capt","Lady","Major","Sir","Countess","Col"], "Unique")
df['Title'].value_counts()
```

```python
#run all 
df['CatAge'] = pd.qcut(df["Age"], q=4, labels=False )
df['CatFare'] = pd.qcut(df["Fare"], q=4, labels=False)
df.head
```

```python
df = df.drop(["Name", "Ticket", "Cabin", "Fare"], axis=1)
df.head()
```

```python

df['Embarked_Q'] = df.Embarked == "Q"
df['Embarked_S'] = df.Embarked == "S"
df['Embarked_C'] = df.Embarked == "C"
df['Title_Mr'] = df.Title == "Mr"
df['Title_Miss'] = df.Title == "Miss"
df['Title_Mrs'] = df.Title == "Mrs"
df['Title_Unique'] = df.Title == "Unique"
df = df.drop(["Embarked"], axis=1)
df.head()
```

```python
df.to_csv('cleaned_data.csv', index=False)

current_directory = os.getcwd()
print(current_directory)
```
